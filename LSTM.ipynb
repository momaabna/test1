```
import numpy as np
from keras.layers import LSTM,Dense
from keras.models import Sequential
from keras.utils.np_utils import to_categorical

def create_model():
    model = Sequential()
    model.add(LSTM(20,
              input_shape=(1200, 1),
              batch_size=10
              ))
    model.add(Dense(10))
    model.compile(loss='mse',optimizer ='adam',metrics=['accuracy'])
    return model
model =create_model()
#train
X = np.load('images.npy')
X =np.reshape(X,[900,1200,1])
Y = to_categorical(np.load('labels.npy'))
model.fit(X,Y,validation_split=0.0,epochs=2000,batch_size=10)


#test
x_test = np.load('test_images.npy')

x_test= np.reshape(x_test,[100,1200,1])
print(x_test.shape)
y_test = to_categorical(np.load('test_labels.npy'))
score=model.evaluate(x_test, y_test,verbose=1,batch_size=10)
print('Test Loss:'+str(score[0])+'Test Acc:'+str(score[1]))
```
